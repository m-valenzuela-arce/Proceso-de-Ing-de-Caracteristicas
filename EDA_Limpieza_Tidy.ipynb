## Acciones basadas en el EDA

```python
#Renombramos los nombres de las columnas que son distintos en ambos datafrasmes para poderlos unir sin problemas
df1 = df1.rename(columns={"feccap": "FECHA CAPTURA", "Movimiento": "TIPO MOVIMIENTO", "OPCION ": "OPCION"})
df2 = df2.rename(columns={"TIPO MOV": "TIPO MOVIMIENTO", "OPCION ": "OPCION"})
```

```python
#Revisamos que los nombres de ambas columnas sean iguales
df1.columns == df2.columns
```

```python
#Unimos ambos dataframes
df_diff = pd.concat([df1,df2])
```

```python
#Renombramos los nombres de algunsa columnasc on la intencion de que sean mas claros respecto a su contenido.
df_diff = df_diff.rename(columns={'CVE PROG': 'CLAVE PROGRAMA', 'NOMBRE ESP': 'NOMBRE ESPECIALIDAD', 'NOMBRE OPC': 'NOMBRE OPCION', 'HORAS TEO': 'HORAS TEORIA', 'HORAS LAB': 'HORAS LABORATORIO', 'EDO 1': 'TIPO INSCRIPCION', 'EDO 2': 'ESTATUS MATERIA', 'CAL ORD': 'CALIFICACION ORDINARIO', 'CAL EXT': 'CALIFICACION EXTRAORDINARIO', 'NUM BAJA': 'NUMERO BAJAS', 'CICLO MAT': 'CICLO MATERIA', 'CVE MAESTRO': 'CLAVE MAESTRO', 'CVE MAESTRO2': 'CLAVE MAESTRO2'})
```

```python
df_diff.info()
```

### Eliminamos los registros repetidos

```python
#Eliminamos registros repetidos
df3 = df_diff.drop_duplicates()
```

```python
df3.info()
```

```python
#Eliminamos algunas columnas que hasta el momento concideramos no relevantes para el analisis.
df3 = df3.drop(["CAMPUS", "DIVISION", "NOMBRE DEPARTAMENTO", "DEPARTAMENTO", "PROGRAMA", "ESPECIALIDAD", "NOMBRE ESPECIALIDAD", "OPCION", "NOMBRE OPCION", "MATERIA", "GRUPO", "CICLO MATERIA", "FECHA CAPTURA"], axis=1)
```

```python
df3.head()
```

### Modificacion de tipos de datos

```python

# Columnas numricas
df_numeric = df3.select_dtypes(include=[np.number])
numeric_cols = df_numeric.columns.values
print("Columnas Numericas:",numeric_cols)

# Columnas no numericas
df_non_numeric = df3.select_dtypes(exclude=[np.number])
non_numeric_cols = df_non_numeric.columns.values
print("Columnas No Numericas:",non_numeric_cols)
```

```python
cols_to_cat = ['CLAVE PROGRAMA','PLAN','ESTATUS','TIPO','NOMBRE MATERIA','TIPO INSCRIPCION','ESTATUS MATERIA']
df3[cols_to_cat] = df3[cols_to_cat].astype('category')
```

```python
df3.info()
```

### Manejo de valores faltantes

```python
# obtiene por variable la cantidad de datos nulos y los transforma a una proporcin del total
serie_pctnull = df3.apply(lambda y: 100*np.mean(y.isnull()))
df_pctnull = pd.DataFrame(serie_pctnull, columns=['% null'])
```

```python
df3.isnull().sum(axis = 0)
```

```python
# Visualizacin  de las posicin de datos perdidos 
fig_msno_mat, ax_msno_mat = plt.subplots(nrows = 1, ncols = 1, figsize = (20,6))
msno.matrix(df3,ax = ax_msno_mat, sparkline=False)
```

```python
#Dada la naturaleza de nuestros datos no procederemos a realizar ningn procedimiento de imputacion.
#En el caso de 'NOMBRE MATERIA' Y 'ESTATUS MATERIA' son solo 1 y 4 datos faltantes, para este caso es masconveniente eliminar los registros.
#En el caso de 'GRUPO' lo indicado, antes de proceder a realizar alguna accion sera hablar con el experto para analizar la razon 
#de los datos faltantes y la reelevancia de estos.

#Eliminamos los datos las filas con datos faltantes ya que son un numero muy pequeo de registros.
df3 = df3.dropna(subset=["ESTATUS MATERIA", "NOMBRE MATERIA"])

#De igual manera, par ael fin de nuestro analisis eliminaremos la columna 'CALIFICACION EXTRAORDINARIO' por lo que no habra necesidad de imputacin.
df3 = df3.drop(['CALIFICACION EXTRAORDINARIO'], axis=1)
```
